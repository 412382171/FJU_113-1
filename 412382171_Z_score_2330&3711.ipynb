```python
url = "https://raw.githubusercontent.com/412382171/FJU_113-1/main/IS.csv"
```

```python
df2 = pd.read_csv(url, index_col=0, encoding='unicode_escape')
```

```python
df2
```

```python
A ='Total current assets'
B ='Total current liabilities'
C ='Total assets'
D ='Total retained earnings'
E ='Earnings before interest and tax'
F ='Total equity'
G ='Total liabilities'
H ='Total operating revenue'
TSMC ='2330 TSMC'
ASEH ='3711 ASEH'
```

```python
X1_TSMC = (df1.loc[TSMC,A] - df1.loc[TSMC,B])/df1.loc[TSMC,C]
X2_TSMC = df1.loc[TSMC,D]/df1.loc[TSMC, C]
X3_TSMC = df2.loc[TSMC,E]/df1.loc[TSMC, C]
X4_TSMC = df1.loc[TSMC,F]/df1.loc[TSMC, G]
X5_TSMC = df2.loc[TSMC,H]/df1.loc[TSMC, C]
```

```python
Z_score_TSMC = 1.2*X1_TSMC + 1.4*X2_TSMC + 3.3*X3_TSMC + 0.6*X4_TSMC +1.0*X5_TSMC
Z_score_TSMC, X1_TSMC, X2_TSMC, X3_TSMC, X4_TSMC, X5_TSMC
```

```python
X1_ASEH = (df1.loc[ASEH,A] - df1.loc[ASEH,B])/df1.loc[ASEH,C]
X2_ASEH = df1.loc[ASEH,D]/df1.loc[ASEH, C]
X3_ASEH = df2.loc[ASEH,E]/df1.loc[ASEH, C]
X4_ASEH = df1.loc[ASEH,F]/df1.loc[ASEH, G]
X5_ASEH = df2.loc[ASEH,H]/df1.loc[ASEH, C]
```

```python
Z_score_ASEH = 1.2*X1_ASEH + 1.4*X2_ASEH + 3.3*X3_ASEH + 0.6*X4_ASEH +1.0*X5_ASEH
Z_score_ASEH, X1_ASEH, X2_ASEH, X3_ASEH, X4_ASEH, X5_ASEH
```
