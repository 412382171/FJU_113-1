```python
url = "https://raw.githubusercontent.com/412382171/FJU_113-1/main/Income%20Statement.csv"
```

```python
df2 = pd.read_csv(url, index_col=0, encoding='unicode_escape')
```

```python
df2
```

```python
A ='Total current assets'
B ='Total current liabilities'
C ='Total assets'
D ='Total retained earnings'
E ='Earnings before interest and tax'
F ='Total equity'
G ='Total liabilities'
H ='Total operating revenue'
TSMC ='2330 TSMC'
UNI ='1216 Uni-President'
```

```python
X1_TSMC = (df1.loc[TSMC,A] - df1.loc[TSMC,B])/df1.loc[TSMC,C]
```

```python
X2_TSMC = df1.loc[TSMC,D]/df1.loc[TSMC, C]
```

```python
X3_TSMC = df2.loc[TSMC,E]/df1.loc[TSMC, C]
```

```python
X4_TSMC = df1.loc[TSMC,F]/df1.loc[TSMC, G]
```

```python
X5_TSMC = df2.loc[TSMC,H]/df1.loc[TSMC, C]
```

```python
Z_score_TSMC = 1.2*X1_TSMC + 1.4*X2_TSMC + 3.3*X3_TSMC + 0.6*X4_TSMC +1.0*X5_TSMC
```

```python
Z_score_TSMC, X1_TSMC, X2_TSMC, X3_TSMC, X4_TSMC, X5_TSMC
```

```python
X1_UNI = (df1.loc[UNI,A] - df1.loc[UNI,B])/df1.loc[UNI,C]
```

```python
X4_UNI = df1.loc[UNI,F]/df1.loc[UNI, G]
```

```python
X5_UNI = df2.loc[UNI,H]/df1.loc[UNI, C]
```
